```python
import geopandas as gpd
import folium
```

```python
r1 = gpd.read_file('ad_to_sur.geojson')
r2 = gpd.read_file('sur_to_ad.geojson')
```

```python
import folium

# Create a map centered at a specific location
m = folium.Map(location=[23.5, 54.5], zoom_start=6, tiles='CartoDB positron')  # Example coordinates for Dubai

# Add routes to the map
folium.GeoJson(r1, name='Route Abu Dhabi to Sur', color = "#6e3d3d").add_to(m)
folium.GeoJson(r2, name='Route Sur to Abu Dhabi', color = "#6e3d3d").add_to(m)

# Define image URLs for each place (replace with your actual image URLs)
places = [
    {
        "location": [24.4997017695181, 54.348572697410766],  # Place 1 coordinates
        "name": "Abu Dhabi",
        "images": []
    },
    {
        "location": [23.067615286832154, 57.66197046167764],  # Place 2 coordinates
        "name": "Saiq",
        "images": [
            "img3.jpg", "img4.jpg", "img5.jpg"
        ]
    },
    {
        "location": [23.61682391338751, 58.371523707180074],  # Place 3 coordinates
        "name": "Muscat",
        "images": [
            "img11.jpg"
        ]
    },
    {
        "location": [23.084628580485752, 59.05800556813313],  # Place 4 coordinates
        "name": "Dibab",
        "images": [
            "img6.jpg", "img21.jpg", "img8.jpg", "img22.jpg"
        ]
    },
    {
        "location": [22.831251149580037, 59.260554203625915],  # Place 5 coordinates
        "name": "Tiwi",
        "images": [
            "img12.jpg"
        ]
    },
    {
        "location": [22.581850718100238, 59.52096955553175],  # Place 5 coordinates
        "name": "Sur",
        "images": [
            "img2.jpg"
        ]
    },
    {
        "location": [22.481720636159288, 58.77809341349808],  # Place 5 coordinates
        "name": "Bidiyah",
        "images": [
            "img19.jpg"
        ]
    },
    {
        "location": [22.9126932203313, 57.53424606054728],  # Place 5 coordinates
        "name": "Nizwa",
        "images": [
            "img1.jpg", "img9.jpg"
        ]
    }
    ]

# Add markers for each place
for place in places:
    # Create the HTML content for the popup with a scrollable div
    images_html = ''.join(
        f'<img src="{img}" width="300" height="200" style="display: block; margin-bottom: 10px;">'
        for img in place['images']
    )
    
    popup_content = f"""
        <div style="max-height: 300px; overflow-y: scroll;">
            <h4>{place['name']}</h4>
            {images_html}
        </div>
    """
    
    # Add a marker with a tooltip and a popup
    marker = folium.Marker(
        location=place['location'],
        tooltip=place['name'],  # Tooltip with the name of the place
        popup=folium.Popup(popup_content, max_width=300)  # Popup with images
    )
    
    # Add the marker to the map
    marker.add_to(m)

# Save the map to an HTML file
m.save("Oman_trip.html")
```
